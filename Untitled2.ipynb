```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from aif360.datasets import CompasDataset
from aif360.metrics import BinaryLabelDatasetMetric
from aif360.algorithms.preprocessing import Reweighing
```

```python
from aif360.datasets import CompasDataset

# Try downloading and loading again
dataset = CompasDataset()
df, _ = dataset.convert_to_dataframe()
df.head()
```

```python
from aif360.metrics import BinaryLabelDatasetMetric

metric = BinaryLabelDatasetMetric(
    dataset,
    unprivileged_groups=[{'race': 0}],
    privileged_groups=[{'race': 1}]
)

print("Disparate Impact:", metric.disparate_impact())
print("Statistical Parity Difference:", metric.statistical_parity_difference())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot risk score distribution by race
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="two_year_recid", hue="race", multiple="stack", palette="Set2", bins=10)
plt.title("Risk Score Distribution by Race")
plt.xlabel("Recidivism (1=Yes, 0=No)")
plt.ylabel("Count")
plt.legend(title="Race", labels=["African-American", "Caucasian"])
plt.show()
```

```python
from aif360.algorithms.preprocessing import Reweighing
from aif360.metrics import BinaryLabelDatasetMetric

# Apply reweighing
RW = Reweighing(unprivileged_groups=[{'race': 0}], privileged_groups=[{'race': 1}])
dataset_transf = RW.fit_transform(dataset)

# Evaluate metrics after reweighing
metric_transf = BinaryLabelDatasetMetric(
    dataset_transf,
    unprivileged_groups=[{'race': 0}],
    privileged_groups=[{'race': 1}]
)

print("Post-Reweighing Disparate Impact:", metric_transf.disparate_impact())
print("Post-Reweighing Statistical Parity Difference:", metric_transf.statistical_parity_difference())
```
